```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read each .csv file and print

import glob

path = r'/home/unic0rn/data_analysis/data_cleaning/uk_used_cars'
all_files = glob.glob(path+"/*.csv") #return all files ending in .csv in our folder

for filename in all_files:
    print('\033[1;30;46m' + filename.rsplit('/', 1)[-1] + ': ' + '\033[0;0m')
    print(pd.read_csv(filename, index_col=0, nrows=0).columns.tolist())
    
```

```python
#merge all tables into one
#take only common columns

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None)
    cols = [c for c in df.columns]
    li.append(df[cols])
    
frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
#add a km per litre column

frame.insert(8, 'kml', frame['mpg']*0.425144)
print(frame)
```

```python
#confirm that no value was left out

sum = 0

for ind in li:
    sum += len(ind)
    
print(sum == len(frame))
```

```python
#confirm that headers were not included in concat

print(frame.nunique())
print(frame['fuelType'].unique())
```
